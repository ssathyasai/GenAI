```python
text = """
Employees receive 12 casual leaves annually.
Employees can work from home twice per week.
Travel expenses are reimbursed within 30 days.
Medical insurance is provided to all employees.
"""
```

```python
from nltk.tokenize import sent_tokenize

#Split into Sentences
chunks=sent_tokenize(text)

for i, chunk in enumerate(chunks):
  print(f'Chunk {i+1}:{chunk}')
```

# ***Group 2 Sentence Per Chunk***

```python
sentences=sent_tokenize(text)
chunk_size=2
chunks=[]

for i in range(0, len(sentences), chunk_size):
  chunk=' '.join(sentences[i: i+chunk_size])
  chunks . append(chunk)

for id, chunk in enumerate(chunks, start=1):
  print(f'\nchunk {id}:')
  print(chunk)
```

```python
!pip install -q sentence-transformers scikit-learn
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize
```

```python


sentences=sent_tokenize(text)

model=SentenceTransformer('all-MiniLM-L6-v2')

embeddings=model.encode(sentences)

threshold=0.5

chunks=[]
current_chunk=[sentences[0]]

for i in range(1, len(sentences)):
  similarity=cosine_similarity([embeddings[i-1]],[embeddings[i]])[0][0]
  if similarity>threshold:
    current_chunk.append(sentences[i])
  else:
    chunks.append(' '.join(current_chunk))
    current_chunk=[sentences[i]]

# Handle the last chunk after the loop finishes
chunks.append(' '.join(current_chunk))

for id,chunk in enumerate(chunks, start=1):
  print(f'\nChunk {id}:')
  print(chunk)
```

## ***LOAD DATA***

```python
chunks = [

    "Employees receive 12 casual leaves annually.",

    "Employees receive 15 sick leaves annually.",

    "Employees may work from home twice per week.",

    "Travel expenses are reimbursed within 30 days.",

    "All employees are covered under company medical insurance."

]

print("Total Chunks:", len(chunks))
```

# **Load Embedding Model**

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer("all-MiniLM-l6-v2")
print("\n Embedding Model Loaded Successfully")
```

# **Generate Embeddings**

```python
embeddings=model.encode(chunks)
print("\n Embeddings Shape")
print(embeddings.shape)
```

# **Create FAISS Index**

```python
!pip install faiss-cpu
```

```python
import  faiss
dimension=embeddings.shape[1]
index=faiss.IndexFlatL2(dimension)
print("FAISS INDEX CREATED SUCCESSFULLY")
```

# **ADD Embeddings to FAISS**

```python
import numpy as np
index.add(
    np.array(embeddings,dtype=np.float32)
)

print("Vectors Stored :",index.ntotal)
```

# **User Query**

```python
query="How many casual leaves are allowed"
print(query)
```

# **Convert Query to Embeddings**

```python
query_embedding=model.encode(
    [query]
)
```

# **Search FAISS**

```python
k=3
distances, indices=index.search(
    np.array(query_embedding,dtype=np.float32),k
)
```

# **Display Results**

```python
print("Matching Chunks\n")
for rank,idx in enumerate(indices[0]):
  print(f'Rank{rank+1}:')
  print(chunks[idx])
  print("Distance:",distances[0][rank])
  print("-"*50)
```
