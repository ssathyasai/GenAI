```python
from reportlab.pdfgen import canvas

pdf_file = "employee_policy.pdf"

c = canvas.Canvas(pdf_file)

content = """
Leave Policy

Employees receive 12 casual leaves annually.
Employees receive 15 sick leaves annually.
Unused casual leaves cannot be carried forward.

Work From Home Policy

Employees may work from home twice per week.
Manager approval is required for additional remote work.

Travel Policy

Travel expenses are reimbursed within 30 days.
Original receipts must be submitted for reimbursement.

Medical Insurance Policy

All employees are covered under company medical insurance.
Dependent coverage is available for spouse and children.
"""

y = 800

for line in content.split("\n"):
    c.drawString(50, y, line)
    y -= 20

c.save()

print("PDF Created Successfully")
```

```python
from PyPDF2 import PdfReader

reader = PdfReader("employee_policy.pdf")

text = ""

for page in reader.pages:
    text += page.extract_text()

print(text)
```

```python
print("Original Document:\n")
print(text)

print("\nTotal Characters:", len(text))
print("Total Words:", len(text.split()))
```

# *Task 2: Fixed Size Chunking*

```python
def fixed_chunk(text, chunk_size):
    chunks = []

    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])

    return chunks
```

```python
for size in [100,200,300]:

    print(f"\n\nChunk Size = {size}")

    chunks = fixed_chunk(text,size)

    for i,chunk in enumerate(chunks):

        print("\nChunk Number:",i+1)
        print(chunk)
        print("Chunk Length:",len(chunk))
```

### **Analysis**
Are sentences being split?

Yes.

Is context preserved?

Partially.

Limitations?

Sentence boundaries are ignored and meaning may be lost.

# ***Task 3: Recursive Chunking***

```python
!pip install langchain-text-splitters -q
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
configs = [
    (200,20),
    (200,50),
    (300,50)
]

for size,overlap in configs:

    print(f"\nChunk Size={size}, Overlap={overlap}")

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=size,
        chunk_overlap=overlap
    )

    chunks = splitter.split_text(text)

    for i,chunk in enumerate(chunks):

        print("\nChunk Number:",i+1)
        print(chunk)
```

## **Analysis**

Overlap keeps some text from the previous chunk.

This prevents loss of context and improves retrieval accuracy.

# ***Task 4: Sentence Based Chunking***

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

for i,sentence in enumerate(sentences):

    print("Sentence Number:",i+1)
    print(sentence)
    print()
```

## ***Analysis***

Sentence chunking preserves meaning better because complete
sentences remain together.

# ***Task 5: Semantic Chunking***

```python
semantic_chunks = {
    "Leave Policy":
    """
    Employees receive 12 casual leaves annually.
    Employees receive 15 sick leaves annually.
    Unused casual leaves cannot be carried forward.
    """,

    "Work From Home Policy":
    """
    Employees may work from home twice per week.
    Manager approval is required for additional remote work.
    """,

    "Travel Policy":
    """
    Travel expenses are reimbursed within 30 days.
    Original receipts must be submitted for reimbursement.
    """,

    "Medical Insurance Policy":
    """
    All employees are covered under company medical insurance.
    Dependent coverage is available for spouse and children.
    """
}
```

```python
for i,(title,chunk) in enumerate(semantic_chunks.items()):

    print(f"\nSemantic Chunk {i+1}")
    print(title)
    print(chunk)
```

### ***Analysis***

Semantic chunking groups related information together.

This improves retrieval because complete topics remain in a single chunk.
